```python
class Meralist:
  def __init__(self):
    self.size=1
    self.n=0

    #create a c type array with size=self.size
    self.A=self.__make_array(self.size)


  def __make_array(self,capacity):
    #create a C type Array (static, referencial )with size ()capactiy
    return (capacity*ctypes.py_object)()


#####create a length funcation
  def __len__(self):
    return self.n

#####create a print function
  def __str__(self):
    result=''

    for i in range(self.n):
      result=result + str(self.A[i]) + ','

    return '[' + result[:-1] + ']'

#####create a indexing funcation

  def __getitem__(self,index):
    if 0<= index <self.n:
      return self.A[index]
    else:
       return 'index Error - index out of range .'

#####create a pop() Function
  def pop(self):
    if self.n==0:
      return 'empty list'

    print(self.A[self.n-1])
    self.n=self.n-1    #reassiged the vallue of "self.n-1" to self.n


#####create a clear funcation

  def clear(self):
    self.n=0
    self.size=1


#####create  a find funcation
  def find(self,item):
    for i in range(self.n):
      if self.A[i]==item:
       return i

    return "value Error - not in list"

#####create a insert() fincation

  def insert(self,pos,item):
    if self.n==self.size:
      self.__resize(self.size*2)

    for i in range(self.n,pos,-1):
      self.A[i]=self.A[i-1]

    self.A[pos]=item
    self.n=self.n+1


######create a del() funcation:  delet the item using index no.

  def __delitem__(self,pos):
    if 0<=pos<self.n:
      for i in range(pos,self.n-1):
        self.A[i]=self.A[i+1]

      self.n=self.n-1

#####create a remove() funcation.
  def remove(self,item):
   pos=self.find(item)

   if type(pos)==int:
    #delet
     self.__delitem__(pos)
   else:
     return pos



#####create a append funcation
  def append(self,item):
    if self.n==self.size:

      self.__resize(self.size*2) #call '__resize()' funcation



    self.A[self.n]=item  #append element
    self.n=self.n+1

  def __resize(self,new_capacity):   #create a' __resize' funcation
    B=self.__make_array(new_capacity)
    self.size=new_capacity


    for i in range(self.n):   #copy content from  A to B
      B[i]=self.A[i]


    self.A=B  # reasiged B to A




#############
#create Array
#len()
#print()
#index()
#pop()
#clear()

#find()
#insert()
#del()
#remove()
#append()


```
